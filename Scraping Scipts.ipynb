```python
import requests
from bs4 import BeautifulSoup
import io
from PyPDF2 import PdfFileReader
import urllib.request
import pandas as pd
```

## Getting Daily Reports
### https://www.health.gov.au/resources/collections/coronavirus-covid-19-at-a-glance-infographic-collection#october-2021

```python
# open the collection page
url = "https://www.health.gov.au/resources/collections/coronavirus-covid-19-at-a-glance-infographic-collection#october-2021"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# get link list to all idv daily report
href_list = set()
date_list = set()
pdf_list = set()
href_list.clear()
date_list.clear()
pdf_list.clear()

mydivs = soup.find("div", {"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
idv_link = mydivs.find_all('a')

file_dir = "Daily Infographics/"
site_path = "http://health.gov.au"
count = 1

for link in (idv_link):
    if count >= 327:
        pdf_href = "http://health.gov.au"+(link.get('href'))
        read = requests.get(pdf_href)
        href_content = read.content
        new_soup = BeautifulSoup(href_content, "html.parser")

        file_path_class = new_soup.find("a", {"class": "health-file__link"})
        issue_date = pdf_href.replace("http://health.gov.au/resources/publications/coronavirus-covid-19-at-a-glance-","").replace("http://health.gov.au/resources/publications/coronavirus-covd-19-at-a-glance-","")
        file_path = file_path_class.get('href')

        print("Report %d: %s"%(count, issue_date))
        response = urllib.request.urlopen(file_path)
        file = open(file_dir+issue_date+".pdf",'wb')
        file.write(response.read())
        file.close()
    count += 1
```

## Getting Vaccination Data
#### Daily Vaccination Dataset from 5 Sep: https://www.health.gov.au/resources/collections/covid-19-vaccination-vaccination-data
#### Daily Vaccination Rollout from 9 Apr: https://www.health.gov.au/resources/collections/covid-19-vaccination-daily-rollout-update
#### Weekly Local Government Area(LGA) vaccination from 28 Aug: https://www.health.gov.au/resources/collections/covid-19-vaccination-geographic-vaccination-rates-lga

#### All vaccination data available on: https://www.health.gov.au/initiatives-and-programs/covid-19-vaccines/numbers-statistics

```python
#Daily Rollout Report(PDF)

url = "https://www.health.gov.au/resources/collections/covid-19-vaccination-daily-rollout-update"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# getting valid stat href
div = soup.find("div",{"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
a_list = div.find_all("a")

count = 1
vaccine_rpt_date = set()
rpt_dir = "Vaccination/Daily Rollout/"
site_path = "http://health.gov.au"

for link in a_list:
    rollout_href = site_path+link.get("href")
    read = requests.get(rollout_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")
    
    file_path_class = new_soup.find("a", {"class": "health-file__link"})
    issue_date = new_soup.find("h1").get_text().lstrip().rstrip().replace('COVID-19 vaccine rollout update – ',"")
    file_path = file_path_class.get('href')
    
    print("Saving Report %d: %s"%(count, issue_date))
    response = urllib.request.urlopen(file_path)
    file = open(rpt_dir+issue_date+".pdf",'wb')
    file.write(response.read())
    file.close()
    count += 1
```

```python
#Daily Dataset(xlsx)

url = "https://www.health.gov.au/resources/collections/covid-19-vaccination-vaccination-data"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# getting valid data href
div = soup.find("div",{"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
a_list = div.find_all("a")

count = 1
vaccine_data_date = set()
daily_national_vaccination = pd.DataFrame()

daily_data_dir = "Vaccination/Dataset(National daily)/Raw Data/"
site_path = "http://health.gov.au"

for link in a_list:
    date_href = site_path+link.get("href")
    read = requests.get(date_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")
    
    file_path_class = new_soup.find("a", {"class": "health-file__link"})
    issue_date = date_href.replace('http://health.gov.au/resources/publications/covid-19-vaccination-vaccination-data-',"")
    file_path = file_path_class.get('href')
    
    df_tmp = pd.read_excel(file_path)
    if count == 1:
        daily_national_vaccination['Measure Name'] = df_tmp['Measure Name']
    daily_national_vaccination[issue_date] = df_tmp['Value']
    
    print("Processing File %d: %s"%(count, issue_date))
    """response = urllib.request.urlopen(file_path)
    file = open(daily_data_dir+issue_date+".xlsx",'wb')
    file.write(response.read())
    file.close()"""
    
    count += 1


file_name = "Vaccination/Dataset(National daily)/Daily national Vaccination(Aggregated).xlsx"
daily_national_vaccination.to_excel(file_name)
daily_national_vaccination
```

```python
#Weekly LGA Dataset(xlsx)

url = "https://www.health.gov.au/resources/collections/covid-19-vaccination-geographic-vaccination-rates-lga"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# getting valid data href
div = soup.find("div",{"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
a_list = div.find_all("a")

count = 1
vaccine_LGA_date = set()
lga_dose_1 = pd.DataFrame()
lga_dose_2 = pd.DataFrame()

weekly_data_dir = "Vaccination/Dataset(LGA weekly)/Raw Data/"
data_filetype="application/vnd.openxmlformats-officedocument.spreadsheetml.sheet"
file_path = ""
site_path = "http://health.gov.au"

for link in a_list:
    date_href = site_path+link.get("href")
    read = requests.get(date_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")
    
    issue_date = date_href.replace('http://health.gov.au/resources/publications/covid-19-vaccination-local-government-area-lga-',"")
    xlsx_path = new_soup.find("a", {"data-filetype": data_filetype}).get("href")
    
    lga_tmp = pd.read_excel(xlsx_path)
    lga_tmp = lga_tmp.iloc[8: , :]
    if count == 1:
        lga_tmp.rename(columns={lga_tmp.columns[0]: "LGA 2019 Name of Residence"}, inplace = True)
        lga_dose_1['LGA 2019 Name of Residence'] = lga_tmp['LGA 2019 Name of Residence']
        lga_dose_1['State'] = lga_tmp['Unnamed: 1']
        lga_dose_1['Remoteness'] = lga_tmp['Unnamed: 2']
        lga_dose_1['Population'] = lga_tmp['Unnamed: 5']
        lga_dose_2 = lga_dose_1.copy()
    
    lga_dose_1[issue_date] = lga_tmp['Unnamed: 3']
    lga_dose_2[issue_date] = lga_tmp['Unnamed: 4']
    
    print("Processing file %d: %s"%(count, issue_date))
    response = urllib.request.urlopen(xlsx_path)
    file = open(weekly_data_dir+issue_date+".xlsx",'wb')
    file.write(response.read())
    file.close()
    count += 1
```

```python
lga_dose_1_new = lga_dose_1[lga_dose_1["State"].str.contains("New South Wales")==True]
file_name_1 = "Vaccination/Dataset(LGA weekly)/LGA Dose 1(Aggregated).xlsx"
lga_dose_1_new.to_excel(file_name_1)
lga_dose_1_new
```
