# Scrape hospital data from Daily Reports
#### https://www.health.gov.au/resources/collections/coronavirus-covid-19-at-a-glance-infographic-collection#october-2021

```python
# open the collection page
url = "https://www.health.gov.au/resources/collections/coronavirus-covid-19-at-a-glance-infographic-collection#october-2021"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# get link list to all idv daily report

dates = []
inpatients = []
ICU = []

mydivs = soup.find("div", {"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
idv_link = mydivs.find_all('a')

site_path = "http://health.gov.au"
case = 1

for link in (idv_link):
    icu_area = []
    inpatient_area = []
    inpatient_data = "0"
    icu_data = "0"
    
    pdf_href = "http://health.gov.au"+(link.get('href'))
    read = requests.get(pdf_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")

    file_path_class = new_soup.find("a", {"class": "health-file__link"})
    issue_date = pdf_href.replace("http://health.gov.au/resources/publications/coronavirus-covid-19-at-a-glance-","").replace("http://health.gov.au/resources/publications/coronavirus-covd-19-at-a-glance-","")
    file_path = file_path_class.get('href')
    

    if issue_date == "04-may-2020":
        case = 2
    elif issue_date == "30-april-2020":
        case = 3

    if case == 1:
        # before 04-may-2020
        icu_area = [211.05,26.7,239.85,195.9]
        inpatient_area = [306.45,29.4,331.65,195.0]

    elif case == 2:
        # before 30-april-2020
        icu_area = [248.985,32.34,273.735,218.46]
        inpatient_area = [345.015,34.32,371.745,220.44]

    elif case == 3:
        # april-2020
        icu_area = [228.897,36.483,261.521,233.282]
        inpatient_area = [336.242,36.483,365.709,233.282]

    try:
        icu_data = tb.read_pdf(file_path, area = icu_area, multiple_tables=False, pandas_options = {'header': None}, pages = '1')[0].iat[0,1]
    except:
        icu_data = "0"

    try:
        inpatient_data = tb.read_pdf(file_path, area = inpatient_area, multiple_tables=False, pandas_options = {'header': None}, pages = '1')[0].iat[0,1]
    except:
        inpatient_data = "0"

    print("%s, Inpatients=%s, ICU=%s"%(issue_date, inpatient_data, icu_data))

    dates.append(issue_date)
    inpatients.append(inpatient_data)
    ICU.append(icu_data)

```

```python
df_hospital = pd.DataFrame({'date': dates, 'inpatients': inpatients, 'icu': ICU}).iloc[::-1]
df_hospital.to_excel("WorkingData/Inpatient Status.xlsx")
df_hospital
```

## Daily Rollout Report(PDF)

```python
url = "https://www.health.gov.au/resources/collections/covid-19-vaccination-daily-rollout-update"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# getting valid stat href
div = soup.find("div",{"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
a_list = div.find_all("a")

count = 1
vaccine_rpt_date = set()
rpt_dir = "Vaccination/Daily Rollout/"
site_path = "http://health.gov.au"

for link in a_list:
    rollout_href = site_path+link.get("href")
    read = requests.get(rollout_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")
    
    file_path_class = new_soup.find("a", {"class": "health-file__link"})
    issue_date = new_soup.find("h1").get_text().lstrip().rstrip().replace('COVID-19 vaccine rollout update – ',"")
    file_path = file_path_class.get('href')
    
    print("Saving Report %d: %s"%(count, issue_date))
    response = urllib.request.urlopen(file_path)
    file = open(rpt_dir+issue_date+".pdf",'wb')
    file.write(response.read())
    file.close()
    count += 1
```

## Weekly LGA Dataset(xlsx)

```python
url = "https://www.health.gov.au/resources/collections/covid-19-vaccination-geographic-vaccination-rates-lga"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

# getting valid data href
div = soup.find("div",{"class": "paragraphs-items paragraphs-items-field-publication-grouping paragraphs-items-field-publication-grouping-full paragraphs-items-full"})
a_list = div.find_all("a")

count = 1
vaccine_LGA_date = set()
lga_dose_1 = pd.DataFrame()
lga_dose_2 = pd.DataFrame()

weekly_data_dir = "Vaccination/Dataset(LGA weekly)/Raw Data/"
data_filetype="application/vnd.openxmlformats-officedocument.spreadsheetml.sheet"
file_path = ""
site_path = "http://health.gov.au"

for link in a_list:
    date_href = site_path+link.get("href")
    read = requests.get(date_href)
    href_content = read.content
    new_soup = BeautifulSoup(href_content, "html.parser")
    
    issue_date = date_href.replace('http://health.gov.au/resources/publications/covid-19-vaccination-local-government-area-lga-',"")
    xlsx_path = new_soup.find("a", {"data-filetype": data_filetype}).get("href")
    
    lga_tmp = pd.read_excel(xlsx_path)
    lga_tmp = lga_tmp.iloc[8: , :]
    if count == 1:
        lga_tmp.rename(columns={lga_tmp.columns[0]: "LGA 2019 Name of Residence"}, inplace = True)
        lga_dose_1['LGA 2019 Name of Residence'] = lga_tmp['LGA 2019 Name of Residence']
        lga_dose_1['State'] = lga_tmp['Unnamed: 1']
        lga_dose_1['Remoteness'] = lga_tmp['Unnamed: 2']
        lga_dose_1['Population'] = lga_tmp['Unnamed: 5']
        lga_dose_2 = lga_dose_1.copy()
    
    lga_dose_1[issue_date] = lga_tmp['Unnamed: 3']
    lga_dose_2[issue_date] = lga_tmp['Unnamed: 4']
    
    print("Processing file %d: %s"%(count, issue_date))
    #response = urllib.request.urlopen(xlsx_path)
    #file = open(weekly_data_dir+issue_date+".xlsx",'wb')
    #file.write(response.read())
    #file.close()
    count += 1
```

### Getting vaccination data from https://www.health.nsw.gov.au/news/pages/2021-nsw-health.aspx

```python
date_list = []
case_list = []
test_list = []
vaccine_list = []
death_list = []

def get_data(link_list):
    
    count = 0
    
    for link in link_list:
        
        if "statistics" in link.get_text():
            href = site_path + link.get("href")
            read_stat = requests.get(href)
            stat_content = read_stat.content
            stat_soup = BeautifulSoup(stat_content,"html.parser")

            num_cases = "No_Data"
            num_deaths = "No_Data"
            num_tests = "No_Data"
            num_vaccines = "No_Data"

            stat_date = stat_soup.find("div", {"class": "newsdate"}).get_text().lstrip().rstrip().strip().replace(" ","-")
            table = stat_soup.find_all("table", {"class": "moh-rteTable-6"})[0]
            tr_list = table.find_all("tr")

            for tr in tr_list:
                text = tr.get_text().lstrip().rstrip().strip()
                num = re.sub('[^0-9]','', text).strip()

                # Case 1: Confirmed Cases
                if "Confirmed cases" in text:
                    num_cases = num

                # Case 2: Total Death
                elif "Deaths" in text:
                    num_deaths = num

                # Case 3: Total Vaccine
                elif "vaccinations" in text:
                    num_vaccines = num
                    
                # Case 4: Total Tests
                elif "Total tests carried out" or "Total persons tested" or "Total" in text:
                    num_tests = num

            date_list.append(stat_date)
            case_list.append(num_cases)
            test_list.append(num_tests)
            death_list.append(num_deaths)
            vaccine_list.append(num_vaccines)
            
            if(num_cases == "No_Data" or num_deaths == "No_Data"or num_tests == "No_Data"):
                print("【ERROR】")

            print("%s: cases %s; tests %s; death %s; vaccine %s"%(stat_date, num_cases, num_tests, num_deaths, num_vaccines))
            
            count += 1
            
            if count > 200:
                break
```

```python
url = "https://www.health.nsw.gov.au/news/pages/2021-nsw-health.aspx"
site_path = "https://health.nsw.gov.au/news/pages/"

read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

ul = soup.find("ul", {"class": "dfwp-column dfwp-list"})
a_list = ul.find_all("a")
get_data(a_list)

url = "https://www.health.nsw.gov.au/news/Pages/2020-nsw-health.aspx"
read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

ul = soup.find("ul", {"class": "dfwp-column dfwp-list"})
a_list = ul.find_all("a")
get_data(a_list)
```

```python
df_statinfo = pd.DataFrame({'date': date_list, 'cases': case_list, 'deaths': death_list, 'tests': test_list, 'vaccines': vaccine_list})
df_statinfo = df_statinfo.iloc[::-1]
#df_statinfo.to_excel("WorkingData/NSW Cases Deaths Tests Vaccines Stats.xlsx")
df_statinfo
```
