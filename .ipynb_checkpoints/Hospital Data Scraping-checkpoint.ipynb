```python
import requests
from bs4 import BeautifulSoup
import io
from PyPDF2 import PdfFileReader
import urllib.request
import pandas as pd
import re
```

```python
url = "https://www.health.nsw.gov.au/news/pages/2021-nsw-health.aspx"
site_path = "https://health.nsw.gov.au/news/pages/"

dates = []
hospitalized = []
ICU = []


read = requests.get(url)
html_content = read.content
soup = BeautifulSoup(html_content,"html.parser")

## find all stat articles
ul = soup.find("ul", {"class": "dfwp-column dfwp-list"})
a_list = ul.find_all("a")

for a in a_list:
    href = site_path + a.get("href")
    
    inpatient = "empty"
    icu = "empty"
    
    ## redirect to the corresponde article and get contents
    stat_read = requests.get(href)
    stat_content = stat_read.content
    stat_soup = BeautifulSoup(stat_content, "html.parser")
    dates.append(stat_date)
    
    ## filter the pages
    stat_date = stat_soup.find("div", {"class": "newsdate"}).get_text().lstrip().rstrip().strip().replace(" ", "-")
    if("COVID-19 cases admitted to hospital" in stat_soup.get_text()):
        p_list = stat_soup.find_all("p")
        for p in p_list:
            string = p.get_text()
            if ("COVID-19 cases admitted to hospital" in string):
                substr = string.replace("There are currently ","").replace(" COVID-19 cases admitted to hospital, with ",";").split(" people in intensive care")[0].replace(",","")
                inpatient = substr.split(";")[0].lstrip().rstrip().strip()
                icu = substr.split(";")[1].lstrip().rstrip().strip()
                break
        
        if inpatient == "empty" or icu == "empty":
            div_list = stat_soup.find_all("div")
            for div in div_list:
                string = div.get_text()
                if ("COVID-19 cases admitted to hospital" in string):
                    substr = string.split("There are currently")[1].split(" people in intensive care")[0].replace(" COVID-19 cases admitted to hospital, with ",";").replace(",","")
                    inpatient = substr.split(";")[0].lstrip().rstrip().strip()
                    icu = substr.split(";")[1].lstrip().rstrip().strip()
                    break
        
        hospitalized.append(inpatient)
        ICU.append(icu)
        print("%s: in hospital-%s; ICU-%s"%(stat_date, inpatient, icu))
    else:
        continue
```
