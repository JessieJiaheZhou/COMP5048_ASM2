```python
import pandas as pd
```

```python
pop1_path = ('NSW Population.xlsx')
df_pop1= pd.read_excel(pop1_path)
df_pop1
```

```python
pop2_path = ('Population by LHD.xlsx')
df_pop2 = pd.read_excel(pop2_path)
df_pop2
```

```python
join_df = pd.merge(df_pop1, 
                   df_pop2,
                   on='LGA', how='left')

join_df
```

```python
filename = 'FinalData/LHD-LGA Mapping.xlsx'
join_df.to_excel(filename)
```
